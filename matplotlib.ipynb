```python
#line graph
import matplotlib.pyplot as plt
x = [1,2,3,4]
y = [4,5,6,7]
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.title("line graph")
plt.plot(x,y)
plt.show()
```

```python
#pie chart
x=[20,30,40,50]
y=["bat","ball","bag","jar"]
A=[0.2,0.3,0,0]
plt.pie(x,labels=y,explode=A)
plt.show()
```

```python
x = [10,20,30,40]
plt.pie(x)
plt.show()
```

```python
X=[1,2,3,4]
Y=[7,8,9,10]
plt.xlabel("X AXIS")
z=['r','blue','pink','green']
plt.bar(X,Y,color=z)
plt.show()
```

```python
x=[2,4,6,8]
y=[1,3,5,7]
z=['r','blue','pink','green']
plt.barh(x,y,color=z)
plt.show()
```

```python
x=[2,4,6,8]
y=[1,3,5,7]
z=['r','blue','pink','green']
sizes=[2000,300,400,500]
plt.scatter(x,y,color=z,marker="*",edgecolors="green",s=sizes)
plt.show()
```

```python
x=[2,4,6,8]
y=[1,3,5,7]
z=['r','blue','pink','green']
sizes=[2000,300,400,500]
plt.xlabel("X axis",fontsize=20)
plt.ylabel("Y axis",fontsize=20)
plt.scatter(x,y,color=z,marker="*",edgecolors="green",s=sizes,alpha=1)
plt.show()
```

```python
x = [10,20,30,40]
y = ["Apple" , "Banana" , "Mange" , "Orange"]
plt.pie(x, labels = y)
plt.show()
```

```python
x = [10,20,30,40]
y = ["Apple" , "Banana" , "Mange" , "Orange"]
z = ["black" , "red" , "lightblue" , "b"]
plt.pie(x, labels = y, colors = z)
plt.show()
```

```python
x = [1,2,3,4,5]
y = [1,2,3,4,5]
z = ["blue","red","black","grey","purple"]
plt.scatter(x,y,color = z)
plt.show()
```

```python
x = [1,2,3,4]
y = [1,2,3,4]
plt.scatter(x,y,color = "black")
plt.show()
```

```python
x = [1,2,3,4,5]
y = [1,2,3,4,5]
z = ["blue","red","black","grey","purple"]
plt.scatter(x,y,color = z,marker = "*")
plt.show()
```

```python
x = [1,2,3,4]
y = [4,5,6,7]
plt.subplot(2,2,1)
plt.plot(x,y)
plt.subplot(2,2,2)

x = [10,20,30,40]
y = ["Apple" , "Banana" , "Mange" , "Orange"]
plt.pie(x, labels = y)
plt.subplot(2,2,3)

x = (["A" , "B" , "C" , "D"])
y = ([3,8,1,10])
plt.bar(x,y)
plt.subplot(2,2,4)

x = [1,2,3,4]
y = [1,2,3,4]
colors = ["blue","red","black","grey"]
sizes = [100,200,300,400]
plt.scatter(x,y,c=colors,s=sizes,alpha=0.8,marker ="*",edgecolor="g")
plt.title("scatter",fontsize=20)
plt.xlabel("x axis",fontsize=20)
plt.ylabel("y axis",fontsize=10)
```
