```python
import matplotlib.pyplot as plt
from wordcloud import WordCloud
from collections import Counter
```

```python
text_data="But the dog would not let the cattle get near the manger, as he wanted it all for himself. So, he snarled and bared his teeth at the cattle. It was as if the manger was filled with the best of meat and bones, which he wished to keep for himself. But the manger really contained only straw."
```

```python
wordcloud=WordCloud(width=800,height=400,background_color='black').generate(text_data)
plt.figure(figsize=(10,5))
plt.imshow(wordcloud)
plt.title('WordCloud')
plt.show()
```

```python
words=text_data.split()
word_counts=Counter(words)
```

```python
plt.figure(figsize=(10,5))
plt.bar(word_counts.keys(),word_counts.values())
plt.title('Histogram of word frequencies')
plt.xlabel('Words')
plt.ylabel('Frequency')
plt.xticks(rotation=90)
plt.show()
```
