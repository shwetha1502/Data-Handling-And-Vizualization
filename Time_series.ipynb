```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data={
    "Date":pd.to_datetime(["2022-01-01","2022-02-01","2023-01-01","2023-02-01"]),
    "Value":[10,15,22,35]
}
df=pd.DataFrame(data)
```

```python
df.set_index("Date",inplace=True)
```

```python
plt.figure(figsize=(10,6))
plt.plot(df.index,df["Value"],marker="X")
plt.xlabel("Date")
plt.ylabel("Value")
plt.title("Time Series with Trend")
plt.grid(True)
plt.show()
```
